```python
def get_channel_subscribers(api_key, channel_id):
    youtube = build("youtube", "v3", developerKey=api_key)
     # Fetch channel statistics
    channel_response = youtube.channels().list(
        part="statistics",
        id=channel_id
    ).execute()
    subscribers = int(channel_response["items"][0]["statistics"]["subscriberCount"])
    
    return subscribers
```

```python
def search_channels_by_name(api_key, channel_name):
   
    # Initialize the YouTube API client
    youtube = build("youtube", "v3", developerKey=api_key)
    
    # Search for channels by name
    search_response = youtube.search().list(
        part="snippet",
        q=channel_name,
        type="channel",
        maxResults=50
    ).execute()
    
    channels = []
    for item in search_response['items']:
        channel_info = {
            "channelId": item["snippet"]["channelId"],
            "title": item["snippet"]["title"],
            "description": item["snippet"]["description"],
            "publishedAt": item["snippet"]["publishedAt"]
        }
        channels.append(channel_info)
    
    return channels

```

```python
# Example usage
api_key = "AIzaSyCV5ylL_bf9-MI9LlWmxknDsuyrF1G0mX8"
channel_name = 'Ti Koze ak TT'  # Replace with the channel name you want to search for

channels = search_channels_by_name(api_key, channel_name)

for channel in channels:
    print(f"Channel ID: {channel['channelId']}")
    print(f"Title: {channel['title']}")
    print(f"Description: {channel['description']}")
    print(f"Published At: {channel['publishedAt']}")
    print("="*40)
```
